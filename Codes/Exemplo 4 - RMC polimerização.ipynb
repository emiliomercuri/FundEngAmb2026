```
dt=.1
Qin=1
k=1
V=1
m_in=1
rhom=.9
rhop=1
rhoin = rhom
Qout = Qin
m_old = m_in
t = 0
```

```
while(t<10):
  m_new = m_old + (Qin*m_in- k*V*m_old - Qout*m_old) * dt/V
  rhoout = rhop - (rhop - rhom) * m_new/m_in
  Qout = Qin * rhoin / rhoout
  m_old = m_new
  print(m_new, Qout, Qout * m_new)
  t = t + dt
```

A solução numérica indica que a concentração do monômero saindo do reator e operando em regime permanente é: 0,513 [adimensional]. (mesmo valor da solução analítica, porém com menos trabalho).

A solução numérica também fornece a vazão no estado estacionário, $Q_{out} = 0.948$ (em algum sistema de unidades).
