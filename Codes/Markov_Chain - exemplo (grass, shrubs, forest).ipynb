Considere 3 estados (de desenvolvimento ecológico) para uma certa área degradada (grass, shrubs, forest), com as seguintes probabilidades:

<img src="https://www.dropbox.com/s/3v00otn5rs2yt2q/cadeia_markov.png?dl=1" width="300">

Vamos considerar

* grass G = h
* shrubs S = a
* forest F = f


A matriz de probabilidades é:

<img src="https://www.dropbox.com/s/2u8ia7v070n6lhl/matriz.png?dl=1" width="300">

$$
\mathbb{P}=
\left(\begin{array}{ccc}
0,5 & 0,45 & 0,05 \\
0,1 & 0,5 & 0,4 \\
0 & 0,1 & 0,9
\end{array}\right)
$$

A Relação entre o estado atual e o estado futuro é:

$$\quad \quad \quad \quad \quad \quad \quad \text{Próximo Estado}$$
$$\quad \quad \quad \quad \quad \quad \quad G \quad \quad S \quad \quad F$$
$$
\text{Estado atual} \quad
\begin{array}{l}
G \\
S \\
F
\end{array} \quad
\left(\begin{array}{ccc}
0,5 & 0,45 & 0,05 \\
0,1 & 0,5 & 0,4 \\
0 & 0,1 & 0,9
\end{array}\right)
$$

```python
import numpy as np
import random as rm
```

```python
# The statespace
states = ["Grass","Shrubs","Forest"]

# Possible sequences of events
transitionName = [["GG","GS","GF"],["SG","SS","SF"],["FG","FS","FF"]]
# A transição Forest -> Grass não é permitida nesse exemplo

# Probabilities matrix (transition matrix)
transitionMatrix = [[0.5, 0.45, 0.05],
                    [0.1, 0.5, 0.4],
                    [0.0, 0.1, 0.9]]
```

```python
# A function that implements the Markov model to forecast the state.
def state_forecast(days):
    # Choose the starting state
    State = "Grass"
    print("Start state: " + State)
    # Shall store the sequence of states taken. So, this only has the starting state for now.
    StateList = [State]
    i = 0
    # To calculate the probability of the StateList
    prob = 1
    while i != days:
        if State == "Grass":
            change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
            if change == "GG":
                prob = prob * 0.5
                StateList.append("Grass")
                pass
            elif change == "GS":
                prob = prob * 0.45
                State = "Shrubs"
                StateList.append("Shrubs")
            else:
                prob = prob * 0.05
                State = "Forest"
                StateList.append("Forest")
        elif State == "Shrubs":
            change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
            if change == "SG":
                prob = prob * 0.1
                State = "Grass"
                StateList.append("Grass")
                pass
            elif change == "SS":
                prob = prob * 0.5
                State = "Shrubs"
                StateList.append("Shrubs")
            else:
                prob = prob * 0.4
                State = "Forest"
                StateList.append("Forest")
        elif State == "Forest":
            change = np.random.choice(transitionName[2],replace=True,p=transitionMatrix[2])
            if change == "FS":
                prob = prob * 0.1
                State = "Shrubs"
                StateList.append("Shrubs")
                pass
            else:
                prob = prob * 0.9
                State = "Forest"
                StateList.append("Forest")
        i += 1
    print("Possible states: " + str(StateList))
    print("End state after "+ str(days) + " days: " + State)
    print("Probability of the possible sequence of states: " + str(prob))

# Function that forecasts the possible state for the next 2 days
state_forecast(2)
```
