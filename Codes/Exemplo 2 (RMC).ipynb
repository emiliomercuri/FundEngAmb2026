```
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Condições iniciais para as componentes A, B, C e D
a0, b0, c0, d0 = 1, 1.5, 0.1, 0.1
# A grid of time points
t = np.linspace(0, 2, 100)

# The model differential equations
def deriv(z, t):
    a, b, c, d = z
    dadt = 1 - 2*a - 4*a*b
    dbdt = 1.5 - b - 4*a*b
    dcdt = 0.1 - c + 4*a*b
    dddt = 0.1 - d + a
    return dadt, dbdt, dcdt, dddt
```

```
# Initial conditions vector
z0 = a0, b0, c0, d0
# Integrate the equations over the time grid, t.
ret = odeint(deriv, z0, t)
a, b, c, d = ret.T

# Plot the data on three separate curves for X(t) and Y(t)
fig = plt.figure()
ax = fig.add_subplot(111)
#fig = plt.figure(facecolor='w')
#ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.set_xlabel('Tempo (adimensional)')
ax.set_ylabel('Concentrações adimensionais')
#ax.set_ylim(0,1.2)
#ax.yaxis.set_tick_params(length=0)
#ax.xaxis.set_tick_params(length=0)
#ax.grid(b=True, which='major', c='gray', lw=0.5, ls='--')
ax.plot(t, a, 'b', alpha=0.5, lw=2, label='A')
ax.plot(t, b, 'r', alpha=0.5, lw=2, label='B')
ax.plot(t, c, 'y', alpha=0.5, lw=2, label='C')
ax.plot(t, d, 'g', alpha=0.5, lw=2, label='D')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
#for spine in ('top', 'right', 'bottom', 'left'):
#    ax.spines[spine].set_visible(False)
plt.show()
```
