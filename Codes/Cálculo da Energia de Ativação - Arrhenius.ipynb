Cálculo da Energia de Ativação para a reação bimolecular do Óxido nítrico com nitrito de cloro para formar dióxido de nitrogênio e cloreto de nitrosilo:

$$
\mathrm{NO}+\mathrm{ClNO}_{2} \rightarrow \mathrm{NO}_{2}+\mathrm{ClNO}
$$

```python
dados = pandas.read_csv('dados-ex1.csv')
```

```python
dados
```

```python
T = dados['T'].values
print(T)
```

```python
# Cálculo da inversa da temperatura
Treciproco = numpy.reciprocal(T)
print(Treciproco)
```

```python
k = dados['k'].values
ln_k = numpy.log(k)
ln_k
```

```python
# Cálculo da regressão linear
from scipy import stats
```

```python
coef_ang, coef_lin, r_corr, p_valor, desv_padrao = stats.linregress(Treciproco,ln_k)
```

```python
print(coef_ang, coef_lin, r_corr)
```

### Gráfico dos dados linearizados e do ajuste de regressão linear

```python
pyplot.figure(figsize=(6,4))
pyplot.plot(Treciproco,ln_k, 'bo')
pyplot.plot(Treciproco, Treciproco * coef_ang + coef_lin, 'r')
#pyplot.ylim(-0.1,1.1)
pyplot.xlabel(r'$T^{-1}$')
pyplot.ylabel(r'$\ln(k)$');
```

```python
T_ativacao = - coef_ang
k0 = numpy.exp(coef_lin)
```
