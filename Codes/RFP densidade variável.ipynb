### Solução numérica do Reator de Fluxo Pistonado com densidade mássica variável

```
# define as variáveis e condições iniciais
# (condições na entrada do reator)
a=1
b=0
u=1
k=1
dz = .0001 
z=0 
PhiAold = u * a
PhiBold = 0
```

```
while (PhiAold > .5):
  PhiAnew = PhiAold - 2 * k * PhiAold**2 / u**2 * dz 
  PhiBnew = PhiBold + k * PhiAold**2 / u**2 * dz
  u = PhiAnew + PhiBnew
  PhiAold = PhiAnew
  PhiBold = PhiBnew
  z = z + dz

print('%.4f'% z)
print('%.4f'% PhiAold)
```
